1 - Como estÃ¡ distribuÃ­do o nÃ­vel de experiÃªncia dos funcionÃ¡rios?

```python
# DistribuiÃ§Ã£o do nÃ­vel de experiÃªncia
experience_distribution = df['experience_level'].value_counts()

# Criando um grÃ¡fico de barras para visualizar a distribuiÃ§Ã£o
plt.figure(figsize=(10, 6))
experience_distribution.plot(kind='bar', color='skyblue')
plt.title('DistribuiÃ§Ã£o do NÃ­vel de ExperiÃªncia dos FuncionÃ¡rios')
plt.xlabel('NÃ­vel de ExperiÃªncia')
plt.ylabel('NÃºmero de FuncionÃ¡rios')
plt.show()
```

2- Existe uma relaÃ§Ã£o entre o tipo de emprego e o salÃ¡rio?

```python
# Criando um boxplot para comparar salÃ¡rios por tipo de emprego
plt.figure(figsize=(16, 8))
sns.boxplot(x='job_title', y='salary_in_usd', data=df)
plt.title('RelaÃ§Ã£o entre Tipo de Emprego e SalÃ¡rio')
plt.xlabel('Tipo de Emprego')
plt.ylabel('SalÃ¡rio em USD')
plt.xticks(rotation=45, ha='right')  # Rotaciona os rÃ³tulos do eixo x para melhor legibilidade
plt.show()
```

ExplicaÃ§Ã£o do grafico:<br>
-Cada caixa representa um tipo de emprego.<br>
-A posiÃ§Ã£o vertical das caixas indica a variabilidade salarial dentro de cada categoria de emprego.<br>
-Se as caixas estÃ£o em diferentes alturas, isso sugere que os salÃ¡rios mÃ©dios ou medianos podem variar entre os tipos de emprego.<br>
-Outliers (pontos fora dos bigodes) podem indicar valores salariais excepcionalmente altos ou baixos para determinados tipos de emprego.<br>
-Comparando vÃ¡rias caixas, vocÃª pode ter uma ideia visual de como os salÃ¡rios estÃ£o distribuÃ­dos em diferentes categorias de emprego.

Resposta. Sim, o salario muda conforme o tipo de emprego.

5 - Quais sÃ£o os tipo de moedas usadas para o pagamento e qual volune delas?

```python
tipos_moedas = df.groupby('salary_currency')['salary'].sum()
maior_para_menor = tipos_moedas.sort_values(ascending=False)
maior_para_menor
```

6 - Qual Ã© a taxa de cÃ¢mbio mÃ©dia para a conversÃ£o de salÃ¡rios para USD?

```python

taxa_cambio = df.groupby('salary_currency')['salary_in_usd'].mean()
taxa_maior_para_menor = taxa_cambio.sort_values(ascending=False)
taxa_maior_para_menor
```

8- Em que localizaÃ§Ãµes a empresa tem presencial?

```python
# Agrupar por 'employee_residence' e calcular a soma de 'remote_ratio' para cada grupo
residencia_remote = df.groupby('employee_residence')['remote_ratio'].sum()

# Converter a sÃ©rie resultante para um DataFrame
result_df = pd.DataFrame({'employee_residence': residencia_remote.index, 'Remote_Ratio_Sum': residencia_remote.values})

# Exibir todas as linhas do DataFrame no Jupyter Notebook
pd.set_option('display.max_rows', None)
display(result_df)
```

8 - Como a distribuiÃ§Ã£o de salÃ¡rios varia entre diferentes localizaÃ§Ãµes?

```python
grouped_data = df.groupby('employee_residence')['salary_in_usd'].mean()
# Crie um grÃ¡fico de barras para visualizar a distribuiÃ§Ã£o
plt.figure(figsize=(12, 6))
sns.barplot(x=grouped_data.index, y=grouped_data.values, hue=grouped_data.index, palette='viridis', legend=False)
# "hue=" permite que vocÃª divida os dados com base em uma variÃ¡vel adicional. O que significa que cada barra serÃ¡ dividida por localizaÃ§Ã£o. Cada cor representa uma localizaÃ§Ã£o diferente.
# "palette=" Esse parÃ¢metro define a paleta de cores a ser usada.
# "legend=False" sem legenda no grafico.
plt.title('DistribuiÃ§Ã£o de SalÃ¡rios por LocalizaÃ§Ã£o')
plt.xlabel('LocalizaÃ§Ã£o')
plt.ylabel('SalÃ¡rio MÃ©dio em USD')
plt.xticks(rotation=45, ha='right')  # Ajusta a rotaÃ§Ã£o dos rÃ³tulos no eixo x para melhor legibilidade
plt.show()
```
