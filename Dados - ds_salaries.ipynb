```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# foi baixado o arquivo: https://www.kaggle.com/datasets/ajjarvis/ds-salaries
# Importo o arquivo "ds_salaries" com o pandas
df = pd.read_csv('ds_salaries.csv')
df
```

1 - Como está distribuído o nível de experiência dos funcionários?

```python
# Distribuição do nível de experiência
experience_distribution = df['experience_level'].value_counts()

# Criando um gráfico de barras para visualizar a distribuição
plt.figure(figsize=(10, 6))
experience_distribution.plot(kind='bar', color='skyblue')
plt.title('Distribuição do Nível de Experiência dos Funcionários')
plt.xlabel('Nível de Experiência')
plt.ylabel('Número de Funcionários')
plt.show()
```

2- Existe uma relação entre o tipo de emprego e o salário?

```python
# Criando um boxplot para comparar salários por tipo de emprego
plt.figure(figsize=(16, 8))
sns.boxplot(x='job_title', y='salary_in_usd', data=df)
plt.title('Relação entre Tipo de Emprego e Salário')
plt.xlabel('Tipo de Emprego')
plt.ylabel('Salário em USD')
plt.xticks(rotation=45, ha='right')  # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.show()
```

Explicação do grafico:<br>
-Cada caixa representa um tipo de emprego.<br>
-A posição vertical das caixas indica a variabilidade salarial dentro de cada categoria de emprego.<br>
-Se as caixas estão em diferentes alturas, isso sugere que os salários médios ou medianos podem variar entre os tipos de emprego.<br>
-Outliers (pontos fora dos bigodes) podem indicar valores salariais excepcionalmente altos ou baixos para determinados tipos de emprego.<br>
-Comparando várias caixas, você pode ter uma ideia visual de como os salários estão distribuídos em diferentes categorias de emprego.

Resposta. Sim, o salario muda conforme o tipo de emprego.

5 - Quais são os tipo de moedas usadas para o pagamento e qual volune delas?

```python
tipos_moedas = df.groupby('salary_currency')['salary'].sum()
maior_para_menor = tipos_moedas.sort_values(ascending=False)
maior_para_menor
```

6 - Qual é a taxa de câmbio média para a conversão de salários para USD?

```python

taxa_cambio = df.groupby('salary_currency')['salary_in_usd'].mean()
taxa_maior_para_menor = taxa_cambio.sort_values(ascending=False)
taxa_maior_para_menor
```

8- Em que localizações a empresa tem presencial?

```python
# Agrupar por 'employee_residence' e calcular a soma de 'remote_ratio' para cada grupo
residencia_remote = df.groupby('employee_residence')['remote_ratio'].sum()

# Converter a série resultante para um DataFrame
result_df = pd.DataFrame({'employee_residence': residencia_remote.index, 'Remote_Ratio_Sum': residencia_remote.values})

# Exibir todas as linhas do DataFrame no Jupyter Notebook
pd.set_option('display.max_rows', None)
display(result_df)
```

8 - Como a distribuição de salários varia entre diferentes localizações?

```python
grouped_data = df.groupby('employee_residence')['salary_in_usd'].mean()
# Crie um gráfico de barras para visualizar a distribuição
plt.figure(figsize=(12, 6))
sns.barplot(x=grouped_data.index, y=grouped_data.values, hue=grouped_data.index, palette='viridis', legend=False)
# "hue=" permite que você divida os dados com base em uma variável adicional. O que significa que cada barra será dividida por localização. Cada cor representa uma localização diferente.
# "palette=" Esse parâmetro define a paleta de cores a ser usada.
# "legend=False" sem legenda no grafico.
plt.title('Distribuição de Salários por Localização')
plt.xlabel('Localização')
plt.ylabel('Salário Médio em USD')
plt.xticks(rotation=45, ha='right')  # Ajusta a rotação dos rótulos no eixo x para melhor legibilidade
plt.show()
```

9 - Existe uma relação entre o tamanho da empresa e a proporção de trabalho remoto?

```python
relacao_trabalho = df.groupby('company_size')['remote_ratio'].mean()
plt.figure(figsize=(12,6))
sns.barplot(x=relacao_trabalho.index, y=relacao_trabalho.values)
plt.title('Relação entre tamanho de empresa com trabalho remoto')
plt.xlabel('Tamanho da empresa')
plt.ylabel('Trabalho remoto')
plt.show
```
